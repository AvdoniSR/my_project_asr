```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
from matplotlib.colors import Normalize

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [100, 200, 150, 300, 250]

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Normalizar los valores para el mapa de colores
norm = Normalize(vmin=min(valores), vmax=max(valores))

# Obtener el colormap inverso
cmap = plt.get_cmap('Blues_r')

# Crear las barras con colores basados en los valores
bars = ax.bar(categorias, valores, color=cmap(norm(valores)))

# Añadir etiquetas con los valores en cada barra
ax.bar_label(bars, labels=[str(v) for v in valores], padding=3)

# Añadir título y etiquetas
ax.set_title('Gráfico de Barras con Colores y Etiquetas')
ax.set_xlabel('Categorías')
ax.set_ylabel('Valores')

# Mostrar el gráfico
plt.show()
```
