## 1. **Cargar y armonizar**  
* Lea los tres archivos.
* Complete los códigos municipales y asegúrese de que todas las geometrías cumplan con la norma EPSG 3116.

```python
root_folder=r"/notebooks/datos/geodata_spatoper"
shp_path=root_folder+r"/municipios_colombia.shp"
df_muni = gpd.read_file(shp_path)
csv_path  = root_folder+"/Evaluaciones_Agropecuarias_Municipales_EVA.csv"
stations_csv_path=root_folder+"/Catalogo_Estaciones_IDEAM.csv"
muni = gpd.read_file(shp_path)
csv_eva = pd.read_csv(csv_path, encoding='utf-8')
csv_eva = csv_eva.rename(columns={'CÓD. MUN.': 'COD_MUNI'})
csv_eva['COD_MUNI'] = csv_eva['COD_MUNI'].astype(str).str.zfill(5)
csv_eva["CÓD. \nDEP."] = csv_eva["CÓD. \nDEP."].astype(str).str.zfill(2)
muni['MPIO_CDPMP'] = muni['MPIO_CDPMP'].astype(str).str.zfill(5)
muni['DPTO_CCDGO'] = muni['DPTO_CCDGO'].astype(str).str.zfill(2)
print("\nCRS:", muni.crs, "\n")              
print("\nVista de Muni_Shape:\n", muni.head())
print("\nVista de EVAs:\n", csv_eva.head())
```

#### Qué se hizo en esta primera parte?
1. Se define root_folder como variable de la ruta principal, ya luego se definepara cada tipo de archivo una variable independiente

## 2. **Enfoque en el año de cultivo**

#### Nota
Sin embargo, el rendimiento se puede calcular con los nuevos datos y comparar que tan diferente queda respecto al promedidado.

```python
cbb_muni_res ["New_rend_cal"] = cbb_muni_res["Producción\n(t)"]/cbb_muni_res["Área Sembrada\n(ha)"]
cbb_muni_res.head()
```

## 4. **Número de estaciones**
* Convertir el archivo CSV de la estación a un GeoDataFrame.
* Unir espacialmente cada estación con su departamento (gpd.sjoin).
* Generar una tabla: número de estaciones por departamento.
* Crear:  
a) Un mapa de coropletas con el número de estaciones por departamento,  
b) Un gráfico de barras horizontales con el mismo número (en orden descendente).


```python

csv_stat_ideam = pd.read_csv(stations_csv_path)

# Se extrae latitud y longitud
csv_stat_ideam[['lat', 'lon']] = csv_stat_ideam['Ubicación']\
    .str.replace(r'[()]', '', regex=True)\
    .str.split(',', expand=True)\
    .astype(float)

# Crear geometría tipo Point (lon, lat)
geometry = [Point(xy) for xy in zip(csv_stat_ideam['lon'], csv_stat_ideam['lat'])]

# Crear el GeoDataFrame
csv_stat_ideam = gpd.GeoDataFrame(csv_stat_ideam, geometry=geometry, crs="EPSG:4326")

# Verificar que sea un GeoDataFrame con geometría válida
print(csv_stat_ideam.head(), "\n")
print(csv_stat_ideam.crs)
```

### ¿Qué se hace en este código?

Se redefine el cvs de las estaciones, se tranforma la columna ubicación, en latitu y longitud, es necesario desglozarse teniendo en cuenta la estructura que trae (Puntos parentesis y tipo de dato)
Se asigna la geometría (Tipo punto) el SRC y finalmente se verifica como queda el GeoDataFrame con print

```python
##depart = muni.dissolve(by="DPTO_CCDGO", as_index=False) --> Se lleva las 92 columnas, aumenta la necesidad de computo para datos que no se van a analizar.

depart = muni.dissolve(
    by="DPTO_CCDGO",
    as_index=False,
    aggfunc={
        "AREA": "sum",
        "LATITUD": "mean",
        "LONGITUD": "mean",
        "Shape_Leng": "sum",
        "Shape_Area": "sum",
        "Codigo_Mun": "first"
    }
)
print(depart.head(), "\n")
print("CRS municipios:", depart.crs, "\n")
print("CRS estaciones:", csv_stat_ideam.crs, "\n")
csv_stat_ideam = csv_stat_ideam.to_crs(depart.crs, "\n")
print("CRS Proyectado", csv_stat_ideam.crs)
```

### ¿Qué se hace en este código?
1. En esta parte, primero es necesario disolver el gdf muni usando el código de departamento para simplificar la operación, y tener los datos sólo por departamento.
2. Se seleccionan algunas columnas (Atributos) relevantes y se define qué ahcer ocn ellos. 
3. Luego se verifica el contenido del gdf
4. También se comprueba el CRS de cada gdf, es importante reproyectarlo (Se hace en la línea 18) para que se pueda hacer la operación

```python
# Realizando la unión
statby_dpto = gpd.sjoin(csv_stat_ideam, depart[['DPTO_CCDGO', 'geometry']], how="left", predicate='intersects')
##print(statby_dpto[['DPTO_CCDGO', 'Nombre', 'geometry']].head()) #--> Lo muestra resumido
print("Tabla de la Unión", statby_dpto.head(), "\n") # --> Lo muestra completo
# Contar estaciones por departamento
num_stat = statby_dpto.groupby('DPTO_CCDGO').size().reset_index(name='num_estaciones')

# Unir GeoDataFrame con el conteo
departs = depart.merge(num_stat, on='DPTO_CCDGO', how='left')


# Creando un DataFrame con códigos únicos y nombres de departamento
dpto_nombres = csv_eva[["CÓD. \nDEP.", "DEPARTAMENTO"]].drop_duplicates()

# Unión (merge) con el GeoDataFrame departamental
departs = departs.merge(
    dpto_nombres,
    left_on="DPTO_CCDGO",
    right_on="CÓD. \nDEP.",
    how="left"
)

# Quitar columna
departs = departs.drop(columns=["CÓD. \nDEP."])

print("Número de Estaciones", num_stat, "\n")
print("GeoDataFrame con Num Estaciones:\n", departs, "\n")
```

### ¿Qué se hace en este código?
1. Se hace una unión espacial (sjoin) entre el GeoDataFrame de estaciones (csv_stat_ideam, puntos) y el de departamentos (depart, polígonos).
2. Se usa el modo left, lo cual conserva todas las estaciones e incorpora el código del departamento que intersecta con su ubicación, El objetivo es asignar a cada estación el departamento al que pertenece, basándose en la geometría
3. Se agrupa por DPTO_CCDGO para contar cuántas estaciones hay en cada departamento.
4. Se hace una unión tipo left para mantener todos los departamentos aunque no tengan estacione, pero se adiciona la columna num_estaciones al GeoDataFrame depart
5. Se crea una tabla auxiliar con los nombres de departamentos únicos desde otro DataFrame (csv_eva), Se hace un merge para asignar el nombre del departamento a cada geometría en departs, es necesario este paso para poder realizar el gráfico de barras.
6. Finalmente se visualiza como van quedando los df y gdf en cada paso. (Se puede notar que El código de departamento 11 queda sin departamento que es Bogotá, despues se le asigna)

```python
# Mapeando 

ax = departs.plot(column='num_estaciones', cmap='YlGnBu', figsize=(10,10),
                     legend=True, edgecolor='grey', linewidth=0.1)
ax.set_title('Cantidad Total de estaciones por Departamento')
ax.set_axis_off()
plt.show()
plt.savefig("figures/stat_dep.png", dpi=300, bbox_inches='tight')
```

### Qué se hace?
Simplmente se utiliza el código de ax para definir la figura y realizar el mapa de coropletas

```python
# Unir GeoDataFrame con el conteo
departs.at[3, 'DEPARTAMENTO'] = 'Bogotá D.C.' # --> El código 11 corresponde a Bogotá pero no se asignó
#dep_num_sort = departs.sort_values('num_estaciones', ascending=True)
dep_num_sort = departs[departs['DEPARTAMENTO'].notna()].sort_values('num_estaciones', ascending=True)

fig, ax = plt.subplots(figsize=(12, 6))

bars = ax.barh(dep_num_sort['DEPARTAMENTO'][::-1], dep_num_sort['num_estaciones'][::-1])

# Añadir etiquetas con bar_label
ax.bar_label(bars, labels=[f'{x:,.0f}' for x in dep_num_sort['num_estaciones'][::-1]], padding=3)
ax.set_xlabel('Total Estaciones')
ax.set_title('Número Total de estaciones meteorológicas por Departamento')
plt.tight_layout()
plt.show()
plt.savefig("figures/bar_stat.png", dpi=300, bbox_inches='tight')
```

### ¿Qué hace este código?
1. Lo primero es que se modifica el gdf, para definir Bogotá D.C. como departamento, ya que el csv de EVAs (De donde se sacaron los Nombres de departamentos) no contenía este valor.
2. Luego se hace la unión con el GeoDataFrame de "departs", aunque ya se hace anteriormente para el mapa de Coropletas, aquí se hace de nuevo para poder ordenarlos de una vez.
3. Se crea la figura y se plotea, se asignan etiquetas para ver de manera más precisa el valor de estaciones de cada departamento.

5. **Producción vs. cobertura**
* Descomponer los municipios en departamentos (sumando la producción agrícola).
* Unir la tabla de recuento de estaciones a esta capa descompuesta.
* Generar un diagrama de dispersión de la producción departamental frente al recuento de estaciones; etiquetar los cuatro casos extremos (alta producción/baja cobertura, etc.).

```python
print(csv_muni_eva.columns.tolist(), "\n")
```

```python

pro_mun_2017_ceb = cebb_muni_2017.groupby("COD_MUNI").agg({
    'DPTO_CCDGO':"first",
    'DEPARTAMENTO':"first",
    'AREA':"sum",
    "MUNICIPIO": "first",
    "Producción\n(t)" :"sum",
    'CULTIVO': "first",
    "Área Sembrada\n(ha)": "sum",
    'AÑO': "first"
}).reset_index()

print(pro_mun_2017_ceb.head())
```

```python
# Crear nuevo GeoDataFrame combinando la geometría
pro_mun_2017_ceb = pro_mun_2017_ceb.merge(
    csv_muni_eva[["COD_MUNI", "geometry"]].drop_duplicates("COD_MUNI"),
    on="COD_MUNI",
    how="left"
)

# Convertirlo en GeoDataFrame
pro_mun_2017_ceb = gpd.GeoDataFrame(pro_mun_2017_ceb, geometry="geometry")

print(type(pro_mun_2017_ceb))

# Disolver por departamentos
pro_dep_2017_ceb = pro_mun_2017_ceb.dissolve(by="DPTO_CCDGO",as_index=False)
print(pro_dep_2017_ceb.head())
```

```python
# Añadiendo el cálculo de número de estaciones a este nuevo gdf

pro_dep_2017_ceb = pro_dep_2017_ceb.merge(
    num_stat[["DPTO_CCDGO", "num_estaciones"]],
    on="DPTO_CCDGO",
    how="left"
)
print(pro_dep_2017_ceb)
```

### ¿Qué hacen los códigos anteriores?
1. Después de comprobar los nombres de los atributos, se reutiliza el df previamente creado "cebb_muni_2017" (Que ya filtraba año y municipio para unificar periodos) para crear un df nuevo, con artributos definidos y qué hacer cone llos. Luego se visualiza.
2. Es necesario recuperar la geometría de estos datos, para ello se hace una unión "merge" con csv_muni_eva que es el gdf que contiene la mayoría de información, se utiliza drop_duplicates para que no traer datos duplciados.
3. Luego se define nuevamente como un GeoDataFrame y se disuelve por departamentos.
4. Finalmente se añade el calculo de la cantidad de estaciones con otra unión, teniendo en cuenta el departamento. En estas uniones sencillas se define "how = left" para que se mantenga la estructura original del primer gdf, sobre todo cuidar la geometría.

```python
# Creando diagrama de dispersión
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(
    pro_dep_2017_ceb["num_estaciones"], 
    pro_dep_2017_ceb["Producción\n(t)"], 
    alpha=0.7
)

# Etiquetas de ejes y título
ax.set_xlabel("Número de Estaciones")
ax.set_ylabel("Producción (t) (2017)")
ax.set_title("Producción Departamental vs Recuento de Estaciones")

# Agregando etiquetas
for i, row in pro_dep_2017_ceb.iterrows():
    ax.text(row["num_estaciones"], row["Producción\n(t)"], row["DEPARTAMENTO"], fontsize=8, alpha=0.7)

plt.tight_layout()
plt.show()
plt.savefig("figures/diag_disp.png", dpi=300, bbox_inches='tight')
```

### Ultimo código:
Finalmente se realiza el diagrama de dispersión y se asignan etiquetas a cada uno de los departamentos.
